```python
## Dilation 
# adds pixels to the boundaries of objects in an image. It is useful for connecting disjoint objects and filling 
# small holes.
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('C:/Users/M.Tech DS006/Desktop/OIP.jfif')

kernel=np.ones((3,3),np.uint8)

dilated=cv2.dilate(img,kernel,iterations=1)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Dilated Image')
plt.imshow(dilated)

plt.show()
```

```python
## Erosion
# Removes pixels on object boundaries. It is useful for removing small noise and separating objexcts that are close together.

img=cv2.imread('C:/Users/M.Tech DS006/Desktop/OIP.jfif')

kernel=np.ones((3,3),np.uint8)

eroded=cv2.erode(img,kernel,iterations=1)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title("Eroded Image")
plt.imshow(eroded)

plt.show()
```

```python
## Opening
# opening is an erosion operation followed by a dilation.
# it is useful for removing small objects from an image while preserving the shape and size of larger objects.

img=cv2.imread('C:/Users/M.Tech DS006/Desktop/OIP.jfif')

kernel=np.ones((4,4),np.uint8)

opened=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Opened Image')
plt.imshow(opened)

plt.show()
```

```python
## Closing
#  closing is a dilation operation followed by an erosion. It is useful for closing small

img=cv2.imread('C:/Users/M.Tech DS006/Desktop/OIP.jfif')

kernel=np.ones((5,5),np.uint8)

closed=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Closed Image')
plt.imshow(closed)

plt.show()
```
