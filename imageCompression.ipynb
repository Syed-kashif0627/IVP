```python
#Lossy Compression using JPEG

import cv2
import matplotlib.pyplot as plt

img=cv2.imread('C:/Users/M.Tech DS006/Desktop/img.jpg')

cv2.imwrite('compressedImage.jpg',img,[cv2.IMWRITE_JPEG_QUALITY,5])

compressedImage=cv2.imread('compressedImage.jpg')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Compressed Image')
plt.imshow(compressedImage)
```

```python
## Lossless Compression using PNG


img=cv2.imread('C:\\Users\\M.Tech DS006\\Desktop\\img.jpg')

cv2.imwrite('compressedPngImage.png',img,[cv2.IMWRITE_PNG_COMPRESSION,9])

compressedImg=cv2.imread('compressedPngImage.png')

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('Compressed Image')
plt.imshow(compressedImg)

plt.show()
```
